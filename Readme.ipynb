# Data Wrangling & Analytics Engine

A modular, object-oriented Python framework designed for end-to-end data sanitization, feature engineering, and interactive visual exploration. This tool is optimized for use within Google Colab environments.

## 🚀 Key Features

### 1. Automated Data Cleaning (`DataWrangler`)
* **Smart Ingestion:** Interactively upload CSV files while automatically converting hidden "junk" null strings (e.g., `?`, `N/A`, `NULL`) into true `NaN` values.
* **Auto-Type Inference:** Automatically scans and converts text-based numerical columns into proper integer/float data types.
* **Data Auditing:** Instantly view dataset dimensions, missing value counts, numerical/categorical feature breakdowns, and a 20-row preview.
* **Flexible Imputation:** Handle missing data using `mean`, `median`, `mode`, or a `constant` target value.
* **Outlier & Duplicate Management:** Identify and flag/drop extreme values using the Interquartile Range (IQR) method, and scrub exact duplicate records.
* **Targeted Deletion:** Prune specific rows or columns interactively.

### 2. Machine Learning Preparation (`DataWrangler`)
* **Numerical Scaling:** Extract and scale continuous features using `minmax`, `standard` (Z-score), or `robust` scaling algorithms.
* **Categorical Encoding:** Extract and encode text factors using `onehot`, `ordinal`, or a special `uniform` (0 to 1 scaled) encoding method.
* **Dataset Fusion:** Recombine processed numerical and categorical frames into a single, ML-ready dataset.

### 3. Interactive Visual Analytics (`InsightVisualizer`)
* **Univariate Subplots:** Generate a rich 3-panel view (Horizontal Box Plot, Sequential Scatter, Histogram) for any numerical feature.
* **Smart Bivariate Charts:** Pass any two columns, and the engine will intelligently detect their data types to render the optimal chart (Scatter + OLS trendline, Grouped Bar, or Box Plot).
* **Categorical Frequency:** Generate bar charts with automatic percentage/ratio text overlays.
* **Granular Custom Charts:** Generate standalone Plotly Pie/Donut charts, grouped Bar charts, and custom bin Histograms.
* **Global Statistical Heatmap:** A unified correlation matrix that intelligently blends math based on variable types:
  * *Numeric vs. Numeric:* Pearson's $r$
  * *Categorical vs. Categorical:* Cramér's $V$
  * *Mixed (Categorical vs. Numeric):* Correlation Ratio (Eta / $\eta$)
